```python
#Scatter Plot BMI and Blood Glucose
import matplotlib.pyplot as plt
x=df['BMI']
y=df['Glucose']
z=df['Outcome']
colors = {0:'red' , 1:'blue'}
plt.scatter(x,y , c=z.apply(lambda x : colors[x]))
plt.xlabel ('BMI')
plt.ylabel ('Glucose')
plt.show()
```
